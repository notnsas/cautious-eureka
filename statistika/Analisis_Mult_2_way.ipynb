```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Example dataset
data = pd.DataFrame({
    'score': [135, 155, 150, 175, 180, 195, 180, 160, 165],
    'teaching_method': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'study_hours': ['Low', 'Medium', 'High', 'Low', 'Medium', 'High', 'Low', 'Medium', 'High']
})

# Two-way ANOVA without interaction
model = smf.ols('score ~ C(teaching_method) + C(study_hours)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

```python
class TwoWay:
  def __init__(self, df):
    self.df = df.to_numpy()

  def hitung(self):
    df = self.df
    xij = np.sum(np.sum(df**2))
    t_2 = np.sum(np.sum(df, axis=1))**2
    kb = df.shape[0] * df.shape[1]

    jkt = xij - t_2 / kb
    print(f"JKT = {xij} - ({t_2}/{kb}) = {jkt}")

    ti_sum = np.sum(np.sum(df, axis=1)**2)
    k = df.shape[1]

    jkb = ti_sum / k - t_2 / kb
    print(f"JKB = ({ti_sum} / {k}) - ({t_2} / {kb}) = {jkb}")

    tj_sum = np.sum(np.sum(df, axis=0)**2)
    # tj_sum = 752125
    b = df.shape[0]

    jkk = (tj_sum / b) - (t_2 / kb)
    print(f"JKK = ({tj_sum} / {b}) - ({t_2} / {kb}) = {jkk}")

    jke = jkt - (jkb + jkk)
    print(f"JKE = {jkt} - ({jkb} + {jkk}) = {jke}")


    db = b - 1
    dk = k - 1
    de = (k - 1)*(b - 1)
    s1_2 = jkb / db
    print(f"s1_2 = {jkb} / {db} = {s1_2}")

    s2_2 = jkk / dk
    print(f"s2_2 = {jkk} / {dk} = {s2_2}")

    s3_2 = jke / de
    print(f"s3_2 = {jke} / {de} = {s3_2}")

    f1 = s1_2 / s3_2
    print(f"f1 = {s1_2} / {s3_2} = {f1}")

    f2 = s2_2 / s3_2
    print(f"f2 = {s2_2} / {s3_2} = {f2}")
```
